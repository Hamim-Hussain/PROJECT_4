```python
import pandas as pd
```

```python
file_path = 'DATA_SET/Raw_dataset.csv'
df = pd.read_csv(file_path)
df.head()
df.columns
```

```python
df["Mother qualification"] = df["Mother\'s qualification"]
df["Father qualification"] = df["Father\'s qualification"]
df["Nationality"] = df["Nacionality"]
df = df[["Marital status","Course","Daytime/evening attendance","Nationality","Mother qualification","Father qualification","Gender","Scholarship holder","International","Age at enrollment","GDP","Target"]]

df.head()
```

```python
df.to_csv('DATA_SET/Cleaned_data.csv',header=True,index = False)
```
